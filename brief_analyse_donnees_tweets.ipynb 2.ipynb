```python
import pandas as pd
import altair as alt
from vaderSentiment_fr.vaderSentiment import SentimentIntensityAnalyzer
from operator import itemgetter, attrgetter
```

```python
path = "tweet_fr_ukraine.csv"
df_tweet_fr = pd.read_csv(path)
```

#### Exploration du jeu de données

```python
df_tweet_fr.head()
```

```python
df_tweet_fr.info()
```

```python
df_tweet_fr.isna().sum()
```

#### Nettoyage et changement des colonnes

##### Création d'une nouvelle colonne user_id à partir des index

```python
df_tweet_fr = df_tweet_fr.reset_index()
df_tweet_fr = df_tweet_fr.rename(columns={'index':'user_id'})
df_tweet_fr.head()
```

##### Récupération du premier # dans la colonne hashtags et nettoyage de la nouvelle colonne

```python
df_tweet_fr['hashtags']=df_tweet_fr['hashtags'].str.split(expand=True)[1]
```

```python
df_tweet_fr.head()
```

```python
df_tweet_fr=df_tweet_fr.replace('\'','',regex=True)
df_tweet_fr = df_tweet_fr.replace('\,','', regex=True)
df_tweet_fr.head()
```

##### Conversion de la colonne 'extractedts' en date puis extraction de l'année, mois, jour...

```python
df_tweet_fr["extractedts"]= pd.to_datetime(df_tweet_fr["extractedts"])
```

```python
#Année
df_tweet_fr['year'] = df_tweet_fr["extractedts"].dt.year

#Mois
df_tweet_fr['month'] = df_tweet_fr['extractedts'].dt.month

#Jour
df_tweet_fr['day'] = df_tweet_fr["extractedts"].dt.day

#heure
df_tweet_fr['hour'] = df_tweet_fr['extractedts'].dt.hour

#minute
df_tweet_fr["minute"] = df_tweet_fr["extractedts"].dt.minute
```

```python
df_tweet_fr.columns
```

```python
df_tweet_fr.head()
```

#### Création des deux sous dataframes

```python
#TABLE USER
df_user = df_tweet_fr.loc[:, ["user_id", "username", "following", "followers"]]

#TABLE TWEET
df_tweet = df_tweet_fr.loc[:, ["hashtags", "retweetcount", "year", "month", "day", "hour", "minute", "user_id"]]
```

```python
df_user.head()
```

```python
df_tweet.head()
```

#### Exportation des tables en csv

```python
df_tweet.to_csv('table_tweet.csv')
df_user.to_csv('table_user.csv')
```

### Analyse des données temporelles

#### Changement de l'index 

```python
df_time = df_tweet_fr.set_index('extractedts')
df_time["hashtags"] = df_time["hashtags"].str.lower()
df_time.head()
```

### Visualisation du nombre de hashtags émis par minute

```python
df_tweet_minute = df_time.resample('T')['hashtags'].count().to_frame().reset_index()
df_tweet_minute
```

```python
alt.Chart(df_tweet_minute).mark_bar().encode(x = 'extractedts', y = 'hashtags', opacity=alt.value(0.2),color=alt.value('red')).interactive()
```

### Visualisation du nb de hashtags produits en fonction du mot

```python
keyword = "guerre"
df_tweet_keyword = df_time.resample('T')['hashtags'].apply(lambda s: s.str.contains(keyword).sum()).to_frame().reset_index()
alt.Chart(df_tweet_keyword).mark_bar().encode(x = 'extractedts', y = "hashtags").properties(width = 800)
```
